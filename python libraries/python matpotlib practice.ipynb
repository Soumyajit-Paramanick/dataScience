```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [3,1,5,2,4]

plt.plot(x,y)

plt.xticks(x)
plt.yticks(y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [3, 1, 5, 2]

plt.plot(x, y)

plt.xticks(x, labels=["python", "c", "c++", "java"])
plt.yticks(x)

plt.show()
```

### Text in Matplotlib Plots - How to Add Text on Plots?

```python
import matplotlib.pyplot as plt1
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt1.plot(x,y)
plt1.text(2,5,"python")
plt1.show()
```

```python
import matplotlib.pyplot as plt1
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt1.plot(x,y)
plt1.text(2,5,"python", fontsize = 15)
plt1.show()
```

```python
import matplotlib.pyplot as plt1
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt1.plot(x,y)
plt1.text(2,5,"python", fontsize = 15, style = "italic")
plt1.show()
```

```python
import matplotlib.pyplot as plt1
x = [1,2,3,4,5]
y = [3,1,5,2,6]

plt1.plot(x,y)
plt1.text(2,5,"python", fontsize = 15, style = "italic", bbox = {"facecolor" : "red"})
plt1.show()
```
